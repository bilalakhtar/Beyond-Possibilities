```python
# Convert employment, internship, and mentorship responses to numeric values
ppdf["Employment Status"] = ppdf["Are you currently employed?"].map({"Yes (Full time)": 1, "No": 0})
ppdf["Internship Interest"] = ppdf["Are you interested in internship or job placement opportunities after the fellowship?"].map({"Yes": 1, "Maybe": 0})
ppdf["Mentorship Interest"] = ppdf["Would you like to be connected to mentors or industry professionals in your field of interest?"].map({"Yes": 1, "Maybe": 0})
```

```python
# Select skill rating columns
skill_columns = [col for col in ppdf.columns if "How would you rate your current proficiency in the following areas?" in col]
```

```python
# Compute mean skill levels
skill_means = df[skill_columns].mean().sort_values()
```

```python
skill_means
```

### Skill Proficiency Analysis

```python
# --- PLOT 1: Skill Proficiency Levels ---
plt.figure(figsize=(10, 6))
sns.barplot(x=skill_means.values, y=skill_means.index, hue=skill_means.index, palette="coolwarm", legend=False)
plt.xlabel("Average Rating (1-5)")
plt.ylabel("Skill Area")
plt.title("Skill Proficiency Levels Among Participants")
plt.show()
```

### Employment Status Analysis

```python
# --- PLOT 2: Employment Status Distribution ---
plt.figure(figsize=(6, 4))
sns.countplot(x=ppdf["Are you currently employed?"], hue=ppdf["Are you currently employed?"], palette="Set2", legend=False)
plt.xlabel("Employment Status")
plt.ylabel("Count")
plt.title("Employment Status of Participants")
plt.xticks(ticks=[0, 1], labels=["Unemployed", "Employed"])
plt.show()
```

### Career Interest & Job Placement Analysis

### Internship and Mentorship Interest

```python
# --- PLOT 3: Interest in Internships & Mentorship ---
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.countplot(x=ppdf["Internship Interest"], ax=axes[0], palette="viridis")
axes[0].set_title("Interest in Internship")
axes[0].set_xticklabels(["No", "Yes"])
sns.countplot(x=ppdf["Mentorship Interest"], ax=axes[1], palette="coolwarm")
axes[1].set_title("Interest in Mentorship")
axes[1].set_xticklabels(["No", "Yes"])
plt.show()
```

### Sentiment Analysis on Suggestions

```python
from textblob import TextBlob

# Assuming you're analyzing the suggestions column for sentiment
ppdf["Sentiment"] = ppdf["Do you have any suggestions or expectations for how the fellowship can best support your career journey?"].astype(str).apply(lambda x: TextBlob(x).sentiment.polarity)

print(ppdf[["Do you have any suggestions or expectations for how the fellowship can best support your career journey?", "Sentiment"]].head())
```

```python
# --- PLOT 4: Sentiment Distribution ---
plt.figure(figsize=(6, 4))
sns.histplot(ppdf["Sentiment"], kde=True, bins=10, color="purple")
plt.xlabel("Sentiment Score")
plt.ylabel("Count")
plt.title("Distribution of Sentiment Scores in Feedback")
plt.show()
```

```python
# --- PLOT 5: Skill Levels by Employment Status ---
plt.figure(figsize=(10, 6))
sns.boxplot(data=ppdf, x="Employment Status", y=ppdf[skill_columns].mean(axis=1), palette="pastel")
plt.xticks(ticks=[0, 1], labels=["Unemployed", "Employed"])
plt.ylabel("Average Skill Rating")
plt.title("Skill Levels by Employment Status")
plt.show()
```
