# Beyond Possibilities - Post Program Feedback

This notebook contains an exploratory data analysis (EDA) of post-program feedback from the Beyond Possibilities initiative. The program aimed to bridge the gap between education and employment for Deaf individuals in Pakistan. This analysis helps in understanding participant feedback, trainer ratings, overall satisfaction, and employment status after the program.

### Install Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the data

```python
# Load dataset (replace 'file.csv' with actual filename)
postpdf = pd.read_csv("Post-Program.csv", parse_dates=["Timestamp"])
```

### Trainer Ratings Distribution

```python
# 1. Trainer Ratings Distribution (Bar Chart)
rating_cols = [col for col in postpdf.columns if "How would you rate" in col]

trainer_ratings = postpdf[rating_cols].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 10))
sns.barplot(x=trainer_ratings.values, y=trainer_ratings.index, palette="Blues_r")
plt.title("Trainer Ratings Distribution")
plt.xlabel("Average Rating")
plt.ylabel("Trainer, Deaf Interpreter and Sign Language Interpreter")
plt.show
```

### Overall Satisfaction Distribution

```python
# 2. Overall Satisfaction Distribution (Histogram)
satisfaction_col = "How much are you satisfied from the fellowship program overall?"

plt.figure(figsize=(8, 6))
sns.histplot(postpdf[satisfaction_col], bins=10, kde=True, color="purple")
plt.title("Overall Satisfaction Distribution")
plt.xlabel("Satisfaction Rating")
```

### Employment Status Breakdown

```python
# 3. Employment Status Breakdown (Pie Chart)
employment_col = "Are you currently employed?"

plt.figure(figsize=(8, 6))
employment_counts = postpdf[employment_col].value_counts()
plt.pie(employment_counts, labels=employment_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Employment Status Breakdown")
```

### Province/State Distribution

```python
# 4. Province/State Distribution (Bar Chart)
province_col = "Province / State:"

province_counts = postpdf[province_col].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=province_counts.index, y=province_counts.values, palette="coolwarm")
plt.title("Province/State Distribution")
plt.xlabel("Province/State")
plt.ylabel("Count")
plt.tick_params(axis='x', rotation=45)
plt.show()
```
